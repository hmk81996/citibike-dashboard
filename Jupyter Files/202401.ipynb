```python
import pandas as pd
from pathlib import Path
```

```python
#store filepaths as a variable
file1 = Path("Resources/202401-citibike-tripdata.csv")
```

```python
#read data files with pandas library
file1_df = pd.read_csv(file1, low_memory=False)
```

```python
file1_df.head()
```

```python
row_count = file1_df.shape[0]
print(row_count)
```

```python
cleaned_file1_df = file1_df.dropna()
cleaned_file1_df.tail()
```

```python
row_count = cleaned_file1_df.shape[0]
print(row_count)
```

```python
sample_202401_df = cleaned_file1_df.sample(frac=0.05) 
```

```python
row_count = sample_202401_df.shape[0]
print(row_count)
```

```python
#export file as csv
sample_202401_df.to_csv("Resources/ny_citibike_202401.csv", index=False, header=True)
```
