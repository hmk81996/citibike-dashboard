```python
import pandas as pd
from pathlib import Path
```

```python
#store filepaths as a variable
file1 = Path("Resources/202407/202407-citibike-tripdata_1.csv")
file2 = Path("Resources/202407/202407-citibike-tripdata_2.csv")
file3 = Path("Resources/202407/202407-citibike-tripdata_3.csv")
file4 = Path("Resources/202407/202407-citibike-tripdata_4.csv")
file5 = Path("Resources/202407/202407-citibike-tripdata_5.csv")
```

```python
#read data files with pandas library
file1_df = pd.read_csv(file1, low_memory=False)
file2_df = pd.read_csv(file2, low_memory=False)
file3_df = pd.read_csv(file3, low_memory=False)
file4_df = pd.read_csv(file4, low_memory=False)
file5_df = pd.read_csv(file5, low_memory=False)
```

```python
file1_df.head()
```

```python
combined_df = pd.concat([file1_df, file2_df, file3_df, file4_df, file5_df], ignore_index=True)
combined_df.tail()
```

```python
cleaned_combined_df = combined_df.dropna()
cleaned_combined_df.tail()
```

```python
row_count = combined_df.shape[0]
print(row_count)
```

```python
row_count = cleaned_combined_df.shape[0]
print(row_count)
```

```python
combined_df = combined_df.loc[:, ~combined_df.columns.str.contains('^Unnamed')]
combined_df.head(50)
```

```python
sample_202407_df = combined_df.sample(frac=0.05) 
```

```python
row_count = sample_202407_df.shape[0]
print(row_count)
```

```python
#export file as csv
sample_202407_df.to_csv("Resources/ny_citibike_202407.csv", index=False, header=True)
```
